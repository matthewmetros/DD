```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', -1)
df = pd.read_csv('Merge_Rules/LeadRules.csv',  dtype=str)

print(df.shape)
df.head(2)
```

```python
# private Map<String, Integer> myMap = new Map<String, Integer>
# {'a' => 1, 'b' => 2, 'c' => 3};
```

```python
d = {}

for index, row in df.iterrows(): 
    heiarchy = row['Heiarchy']
    if heiarchy in d:
        d[row['Heiarchy']].append(row['Label'])
    else:
        d[row['Heiarchy']] = [row['Label']]
```

```python
print("Map<String, List<String>> m = new Map<String, List<String>>{ ")
for key in d.keys():
    print("'"+key +"'" + " => new Map<String, list<String>>{ ")
    for value in d.get(key):
        print("'"+value+"', ")
```
